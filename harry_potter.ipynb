```python
import gensim
import os
from nltk import sent_tokenize, word_tokenize
from gensim.utils import simple_preprocess
from gensim.models import Word2Vec
```

```python
from gensim.parsing.preprocessing import remove_stopwords,preprocess_string,strip_tags,strip_punctuation,strip_multiple_whitespaces,strip_numeric,strip_short,lower_to_unicode
```

```python
filters=[strip_tags,strip_punctuation,strip_multiple_whitespaces,strip_numeric,remove_stopwords,strip_short,lower_to_unicode]
story=[]
for filename in os.listdir('data'):
  f=open(os.path.join('data',filename),'r')
  text=f.read()
  sent=sent_tokenize(text)
  for i in sent:
    story.append(simple_preprocess(remove_stopwords(i)))
```

```python
model=Word2Vec(min_count=2,vector_size=200,window=5,workers=4)
model.build_vocab(story)
model.train(story,total_examples=model.corpus_count,epochs=10)
```

```python
print(model.wv.index_to_key)
```
